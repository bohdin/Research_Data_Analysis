## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## html table

```python
data = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8')
```

```python
df = data[12]
df
```

```python
df.head()
```

```python
df.shape
```

```python
df.replace('—', np.NAN, inplace=True)
```

```python
df.dtypes
```

```python
df[['2014', '2019']] = df[['2014', '2019']].astype('float64') 
```

```python
df.isnull().sum()/df.shape[0]
```

```python
df.drop(27, inplace=True)
```

```python
df.fillna(df.drop(columns=['Регіон']).mean(), inplace=True)
df
```

```python
df[df['2019'] > df['2019'].mean()]['Регіон']
```

```python
df.iloc[df['2014'].idxmax()]
```

## Visualization 

```python
plt.bar(df['Регіон'], df['2019'], color = [i for i in plt.cm.tab20.colors[:27]])
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
kyiv = df[df['Регіон'] == 'Київ']
kyiv
```

```python

plt.plot(kyiv.iloc[0, 1:].keys(), kyiv.iloc[0, 1:].values)
plt.grid()
plt.show()
```

```python
odesa = df[df['Регіон'] == 'Одеська']
odesa
```

```python
sns.lineplot(x = odesa.columns[1:], y = odesa.values[0, 1:])
plt.title('Odesa')
plt.show()
```

```python
plt.figure(figsize=(10, 9))
plt.pie(
    df['1950'],
    labels=df['Регіон'],
    autopct='%.1f%%',
    pctdistance=0.9,
    labeldistance=1.05
)
plt.show()
```
